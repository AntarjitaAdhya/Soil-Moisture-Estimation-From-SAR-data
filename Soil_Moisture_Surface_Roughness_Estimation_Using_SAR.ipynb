```python
import rasterio
import numpy as np

# Step 1: Read SAR data from the raster file
def read_sar_data(file_path):
    with rasterio.open(file_path) as src:
        sar_data = src.read(1)  # Assuming single-band data
    return sar_data

# Step 2: Compute RMS Height
def compute_rms_height(sar_data):
    height_squared = np.square(sar_data)
    mean_height_squared = np.mean(height_squared)
    rms_height = np.sqrt(mean_height_squared)
    return rms_height

# Step 3: Soil Moisture Estimation
def estimate_soil_moisture(rms_height):
    # Example empirical relationship: Soil moisture is inversely proportional to RMS height
    # You can replace this with a more sophisticated model based on your data and domain knowledge
    soil_moisture = 1 / rms_height
    return soil_moisture

# Main function
if __name__ == "__main__":
    # Provide the file path to your SAR data file
    file_path = "/content/Sentinel1_2019.tif"

    # Step 1: Read SAR data
    sar_data = read_sar_data(file_path)

    # Step 2: Compute RMS Height
    rms_height = compute_rms_height(sar_data)

    # Step 3: Soil Moisture Estimation
    estimated_soil_moisture = estimate_soil_moisture(rms_height)

    # Print the estimated soil moisture
    print("Estimated Soil Moisture:", estimated_soil_moisture)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio

def power_law_spectrum(image):
    """
    Calculate the power-law spectrum of an image.

    Parameters:
    - image: 2D numpy array representing the image

    Returns:
    - Frequencies: 1D numpy array containing the spatial frequencies
    - Spectrum: 1D numpy array containing the power spectrum
    """
    # Perform 2D Fourier transform
    f_transform = np.fft.fft2(image)

    # Shift the zero-frequency component to the center
    f_transform_shifted = np.fft.fftshift(f_transform)

    # Calculate the power spectrum
    spectrum = np.abs(f_transform_shifted)**2

    # Calculate the spatial frequencies
    rows, cols = image.shape
    freq_rows = np.fft.fftfreq(rows)
    freq_cols = np.fft.fftfreq(cols)
    freq_rows_shifted = np.fft.fftshift(freq_rows)
    freq_cols_shifted = np.fft.fftshift(freq_cols)

    return freq_rows_shifted, freq_cols_shifted, spectrum

def plot_power_spectrum(freq_rows, freq_cols, spectrum):
    """
    Plot the power spectrum.

    Parameters:
    - freq_rows: 1D numpy array containing the shifted row frequencies
    - freq_cols: 1D numpy array containing the shifted column frequencies
    - spectrum: 2D numpy array containing the power spectrum
    """
    plt.figure(figsize=(8, 6))
    plt.imshow(np.log(spectrum), extent=(freq_cols.min(), freq_cols.max(), freq_rows.min(), freq_rows.max()), cmap='jet')
    plt.colorbar(label='log(Power Spectrum)')
    plt.xlabel('Spatial Frequency (cycles/pixel)')
    plt.ylabel('Spatial Frequency (cycles/pixel)')
    plt.title('Power Spectrum')
    plt.show()

# Load Sentinel-1 SAR data
file_path = "/content/Sentinel1_2018.tif"
with rasterio.open(file_path) as src:
    sar_data = src.read(1)  # Assuming single-band data, adjust if necessary

# Calculate the power-law spectrum
freq_rows, freq_cols, spectrum = power_law_spectrum(sar_data)

# Plot the power spectrum
plot_power_spectrum(freq_rows, freq_cols, spectrum)
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Define file path
file_path = '/content/Sentinel1_2019.tif'

try:
    # Read Sentinel-1 data
    with rasterio.open(file_path) as src:
        vv_data = src.read(1).astype(float)  # VV band
        vh_data = src.read(2).astype(float)  # VH band

    # Define epsilon value to avoid division by zero
    epsilon = 1e-6

    # Calculate Soil Moisture Index (SMI)
    SMI = np.where((vv_data + vh_data) == 0, np.nan, 10 * np.log10((vv_data - vh_data) / (vv_data + vh_data + epsilon)))

    # Display the SMI values
    print("Soil Moisture Index (SMI) values:")
    print(SMI)

    # Plot the SMI values
    plt.figure(figsize=(8, 6))
    plt.imshow(SMI, cmap='jet', vmin=-30, vmax=0)  # Adjust vmin and vmax as needed
    plt.colorbar(label='SMI (dB)')
    plt.title('Soil Moisture Index (SMI)')
    plt.xlabel('Column')
    plt.ylabel('Row')
    plt.grid(False)
    plt.show()

except Exception as e:
    print("An error occurred:", e)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from scipy.signal import welch

# Step 1: Read SAR data from TIFF file
def read_sar_data(file_path):
    with rasterio.open(file_path) as src:
        sar_data = src.read(1)  # Assuming single-band data, adjust if necessary
    return sar_data

# Step 2: Compute Power Spectral Density (PSD)
def compute_psd(image):
    frequencies, psd = welch(image, fs=1.0, nperseg=1024)
    # Average PSD across all frequency bins
    avg_psd = np.mean(psd, axis=0)
    return frequencies, avg_psd

# Step 3: Estimate soil moisture from PSD
def estimate_soil_moisture(psd):
    # In this example, we'll use a simple method where soil moisture is inversely proportional to the PSD
    # You can replace this with a more sophisticated algorithm if needed
    soil_moisture = 1 / psd
    return soil_moisture

# Step 4: Estimate roughness from PSD
def estimate_roughness(psd):
    # In this example, we'll use a simple method where roughness is directly proportional to the PSD
    # You can replace this with a more sophisticated algorithm if needed
    roughness = psd
    return roughness

# Step 5: Visualize PSD and estimation results
def visualize_results(title, xlabel, ylabel, frequencies, values):
    plt.figure(figsize=(10, 6))
    plt.semilogx(frequencies, values)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

# Main function
if __name__ == "__main__":
    # Step 1: Read SAR data
    file_path = "/content/Sentinel1_2017.tif"
    sar_data = read_sar_data(file_path)

    # Step 2: Compute PSD
    frequencies, psd = compute_psd(sar_data)

    # Step 3: Estimate soil moisture and roughness
    estimated_soil_moisture = estimate_soil_moisture(psd)
    estimated_roughness = estimate_roughness(psd)

    # Step 4: Visualize PSD and estimation results for soil moisture
    visualize_results('Estimated Soil Moisture', 'Frequency (Hz)', 'Soil Moisture', frequencies, estimated_soil_moisture)

    # Step 5: Visualize PSD and estimation results for roughness
    visualize_results('Estimated Roughness', 'Frequency (Hz)', 'Roughness', frequencies, estimated_roughness)

    # Step 6: Print estimated values and their average
    print("Estimated Soil Moisture:")
    print(estimated_soil_moisture)
    avg_soil_moisture = np.mean(estimated_soil_moisture)
    print("Average Soil Moisture:", avg_soil_moisture)

    print("\nEstimated Roughness:")
    print(estimated_roughness)
    avg_roughness = np.mean(estimated_roughness)
    print("Average Roughness:", avg_roughness)
```

```python
import rasterio
import numpy as np
from scipy.signal import convolve
import matplotlib.pyplot as plt

# Read Sentinel-1 data
with rasterio.open('/content/Sentinel1_2018.tif') as src:
    sar_data = src.read(1)  # assuming single-band data

# Perform preprocessing if necessary (e.g., speckle filtering)

# Calculate the local variance as a measure of roughness
kernel = np.ones((3, 3))
mean = convolve(sar_data, kernel, mode='same') / 9
variance = convolve((sar_data - mean)**2, kernel, mode='same') / 9

# Estimate roughness using the fractal method
roughness = np.log(variance) / np.log(2)

# Print the calculated roughness values
print("Roughness values:")
print(roughness)

# Plot the estimated roughness
plt.figure(figsize=(8, 6))
plt.imshow(roughness, cmap='jet')
plt.colorbar(label='Roughness')
plt.title('Estimated Soil Roughness')
plt.xlabel('Column')
plt.ylabel('Row')
plt.grid(False)
plt.show()
```

```python
import rasterio
import numpy as np
from scipy.ndimage import generic_filter
import matplotlib.pyplot as plt

def calculate_cv(arr):
    return np.std(arr) / np.mean(arr)

# Define file paths for each year's SAR data
file_paths = [f"/content/Sentinel1_{year}.tif" for year in range(2014, 2024)]

# Function to process SAR data and calculate soil moisture
def process_sar_data(file_path):
    with rasterio.open(file_path) as src:
        sentinel1_data = src.read()

    # Convert the data from dB to linear scale if needed
    sentinel1_data_linear = 10 ** (sentinel1_data / 10)

    # Apply speckle filtering
    # You can use any speckle filtering method suitable for your data
    # For example, Lee filter
    from skimage.restoration import denoise_nl_means
    speckle_filtered_data = denoise_nl_means(sentinel1_data_linear, h=0.8, fast_mode=False, patch_size=5, patch_distance=3, multichannel=False)

    # Calculate Coefficient of Variation (CV) for soil moisture
    window_size = 5  # Adjust the window size based on your data and resolution
    cv_data = generic_filter(speckle_filtered_data, calculate_cv, size=window_size)

    return cv_data

# Process SAR data for each year
soil_moisture_data = [process_sar_data(file_path) for file_path in file_paths]

# Calculate the average soil moisture for each year
avg_soil_moisture = [np.mean(data, axis=1) for data in soil_moisture_data]

# Extract the first element of each array in avg_soil_moisture
avg_soil_moisture_scalar = [x[0] for x in avg_soil_moisture]

# Plot the comparison between the years using a bar plot
plt.figure(figsize=(10, 6))
years = list(range(2014, 2024))

for i in range(len(years)):
    # Check if all elements in the array are NaN
    if not np.all(np.isnan(avg_soil_moisture_scalar[i])):
        plt.bar(str(years[i]), avg_soil_moisture_scalar[i], label=str(years[i]))

plt.xlabel('Year')
plt.ylabel('Average Soil Moisture (Coefficient of Variation)')
plt.title('Average Soil Moisture Comparison from 2014 to 2023')
plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_fractal_roughness(rms_height, correlation_length, LAI):
    """
    Compute fractal roughness based on the given parameters.

    Parameters:
    - rms_height: Root Mean Square (RMS) height of the surface (in cm)
    - correlation_length: Correlation length of the surface (in cm)
    - LAI: Leaf Area Index, indicating vegetation growth

    Returns:
    - fractal_roughness: Fractal roughness estimate
    """
    # Compute fractal roughness based on the given formula or methodology
    # This may involve different calculations based on the methodology described in the paper
    fractal_roughness = rms_height * np.exp(-LAI / correlation_length)

    return fractal_roughness

def compute_inverted_roughness(rms_height, correlation_length):
    """
    Compute inverted single-scale roughness based on the given parameters.

    Parameters:
    - rms_height: Root Mean Square (RMS) height of the surface (in cm)
    - correlation_length: Correlation length of the surface (in cm)

    Returns:
    - inverted_roughness: Inverted single-scale roughness estimate
    """
    # Compute inverted single-scale roughness based on the given formula or methodology
    # This may involve different calculations based on the methodology described in the paper
    inverted_roughness = rms_height * correlation_length

    return inverted_roughness

# Define the time series data for YA3 and YA5 stations
# Assuming the data format is [year, rms_height, correlation_length, LAI]
data_2014 = np.array([[2014, 15.183243, 0.57, 0.2], [2015, 16.199472, 0.60, 0.3], [2016, 15.108516, 0.65, 0.4], [2017, 16.427942, 0.70, 0.5], [2018, 17.128516, 0.68, 0.6], [2019, 17.792843, 0.63, 0.7]])

# Compute fractal and inverted roughness for each year
fractal_roughness = compute_fractal_roughness(data_2014[:, 1], data_2014[:, 2], data_2014[:, 3])
inverted_roughness = compute_inverted_roughness(data_2014[:, 1], data_2014[:, 2])

# Plot the time series comparison
plt.figure(figsize=(12, 6))

plt.plot(data_2014[:, 0], inverted_roughness, label='Inverted Roughness')
plt.plot(data_2014[:, 0], fractal_roughness, label='Fractal Roughness')
plt.xlabel('Year')
plt.ylabel('Roughness Estimate')
plt.title('Roughness Comparison (2014-2019)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute roughness using PSD-based method
def compute_roughness_psd(psd):
    # Roughness is directly proportional to PSD
    roughness_psd = psd
    return roughness_psd

# Function to compute roughness using Fractal Analysis method
def compute_roughness_fractal(rms_height, correlation_length, LAI):
    # Roughness based on fractal analysis
    roughness_fractal = rms_height * np.exp(-LAI / correlation_length)
    return roughness_fractal

# Sample data for the years 2014 to 2023
years = range(2014, 2024)
rms_height = np.random.uniform(10, 20, size=len(years))
correlation_length = np.random.uniform(0.5, 0.7, size=len(years))
LAI = np.random.uniform(0.1, 0.5, size=len(years))

# Compute roughness using both methods for each year
roughness_psd = compute_roughness_psd(rms_height)
roughness_fractal = compute_roughness_fractal(rms_height, correlation_length, LAI)

# Plotting the comparison
plt.figure(figsize=(10, 6))
plt.plot(years, roughness_psd, marker='o', label='PSD-based Roughness')
plt.plot(years, roughness_fractal, marker='o', label='Fractal Analysis Roughness')
plt.xlabel('Year')
plt.ylabel('Roughness')
plt.title('Comparison of Roughness Estimates (2014-2023)')
plt.legend()
plt.grid(True)
plt.show()
```
